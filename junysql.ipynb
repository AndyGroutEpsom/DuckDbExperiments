```python
%sql select strptime(date, '%b%Y') as dt, \
          "U.K. Investment Grade Bond Index Fund" as raw, \
          "U.K. Investment Grade Bond Index Fund"/100+1 as amount, \
          SUM(amount) OVER (ORDER BY dt) AS cumulative_amount, \
          (EXP(SUM(LN(amount)) OVER (ORDER BY dt))-1)*100 AS cum \
      from 'performance.csv' \
      where date like '%2022'
```

```python
import os

# Open the file in read mode
with open('performance.csv', 'r') as file:
    # Read all lines from the file
    lines = file.readlines()

# Replace spaces with underscores in the first line
lines[0] = lines[0].replace('%', '')
lines[0] = lines[0].replace(' ', '')
lines[0] = lines[0].replace('®', '')

# Open the file in write mode to save the changes
new_file = 'performance_tmp2.csv'
if os.path.exists(new_file):
    os.remove(new_file)
with open(new_file, 'w') as file:
    # Write the modified lines back to the file
    file.writelines(lines)
```
